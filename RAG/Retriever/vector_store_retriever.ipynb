```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_chroma import Chroma
from dotenv import load_dotenv
from langchain_community.document_loaders import PyPDFLoader

load_dotenv()
```

```python
# load the pdf file via document loader
pdf_load = PyPDFLoader(
    file_path='../files/Digital Campus - Black Book.pdf'
)

doc = pdf_load.load()
```

```python
embeddings = GoogleGenerativeAIEmbeddings(model='models/embedding-001')

vector_store = Chroma(
    embedding_function=embeddings,
    persist_directory='../Vectors/Database/'
)
```

```python
# adding documents
vector_store.add_documents(doc)
```

```python
# convert vectorstore into a 
# fetching top 5 results
retriever = vector_store.as_retriever(search_kwargs={"k": 5})
```

```python
query = "what is the tech stack used?"
result = retriever.invoke(query)
```

```python
i = 1
for result in result:
    print("Output ", i, "\n", result.page_content)
    print("metadata ", i, "- ", result.metadata)
    print("\n")
    i = i +1
```
