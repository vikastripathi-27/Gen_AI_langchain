```python
import requests
import os
from dotenv import load_dotenv

load_dotenv()
cricbuzz_api_key =  os.getenv("CRICBUZZ_API_TOKEN")

url = "https://cricbuzz-cricket.p.rapidapi.com/series/v1/international"

headers = {
	"x-rapidapi-key": cricbuzz_api_key,
	"x-rapidapi-host": "cricbuzz-cricket.p.rapidapi.com"
}
```

```python
response = requests.get(url, headers=headers)

result = response.json()
```

```python
print(result)
```

```python
from datetime import datetime

data = result  # your provided dictionary

# Access each month and its series
for month_data in data['seriesMapProto']:
    print(f"Month: {month_data['date']}")
    
    # Access each series in that month
    for series in month_data['series']:
        print(f"  Series ID: {series['id']}")
        print(f"  Name: {series['name']}")
        start_dt = datetime.fromtimestamp(int(series['startDt']) / 1000)
        end_dt = datetime.fromtimestamp(int(series['endDt']) / 1000)
        print(f"  Start: {start_dt.strftime('%Y-%m-%d')}")
        print(f"  End: {end_dt.strftime('%Y-%m-%d')}")
        print("---")
```
